```python
import pandas as pd
import numpy as np
```

```python
mean_pop = 32.33
threshold = 2

triptimes = pd.read_csv("https://github.com/fedhere/DSPS_FBianco/raw/refs/heads/main/data/triptimes.csv", header = None)
triptimes
```

```python
mean_time = triptimes[0].mean()
std_time = triptimes[0].std()
n = triptimes.shape[0]
var_time = std_time/np.sqrt(n)

Z = (mean_time-mean_pop)/var_time
print(Z)
if Z > threshold:
  print("Null hypothesis is rejected!")
else:
  print("Null hypothesis is not rejected!")
```
